# III- Repondre a des question a partir de Dataset


```python
sns.countplot(x='type',data=df,ec='black',lw=1)
plt.title('Type',fontsize=18)
plt.xlabel("type")
plt.ylabel("nombre")
plt.show()
```

***Q: Quel top 5 pays produit le plus de contenu sur Netflix?***


```python
plt.figure(figsize=(12,10))
sns.countplot(y='country',order=df['country'].value_counts().index[0:5],data=df,ec='black',lw=2)
plt.title('Contenu relatif au pays sur Netflix.',fontsize=30)
plt.ylabel('Pays',fontsize=13)
plt.xlabel('Production',fontsize=13)
plt.show()
```

***Q : En quelle année Netflix a-t-il généré le plus de contenu?***

```python
sns.displot(df["release_year"],color="red")
plt.title("Production par ans",fontsize=17)
plt.ylabel('Production',fontsize=13)
plt.xlabel('Année',fontsize=13)
plt.show()
```

***Q : Que sont les top 10 acteurs***

```python
top_realisateur = df['cast'].value_counts().head(10)
plt.figure(figsize=(5, 5))
top_realisateur.plot(kind='bar', color='blue')
plt.xlabel('réalisateurs')
plt.xticks(rotation=60,ha='right')
plt.title('Top 10 Acteurs')
plt.tight_layout()
plt.show()
```

***Q : Quelles sont les top 5 catégories***

```python
top_categories = df['listed_in'].value_counts().head()
plt.figure(figsize=(6, 6))
plt.barh(top_categories.index, top_categories.values,edgecolor='black', linewidth=2)
plt.title("catégorie", fontsize=18)
plt.xlabel("", fontsize=13)
plt.show()
```

***Q : Que Sont TOP 5 rélisateur***

```python
plt.figure(figsize=(10,10))
plt.pie(list(df['director'].value_counts().head()),labels=list(df['director'].value_counts().head().keys()),autopct='%0.1f%%')
plt.title("TOP 5 rélisateur")
plt.show()
```

***Q : Identifier des contenus similaires en faisant correspondre des caractéristiques textuelle***

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

descriptions = df['description']


vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(descriptions)
cosine_sim = cosine_similarity(tfidf_matrix)

content_index = 0

similar_contents = list(enumerate(cosine_sim[content_index]))
similar_contents = sorted(similar_contents, key=lambda x: x[1], reverse=True)

titles = [df['title'].iloc[content[0]] for content in similar_contents[1:6]]

scores = [content[1] for content in similar_contents[1:6]]

plt.bar(titles, scores)
plt.ylabel('Similarity Score')
plt.title('Top 5 Similar Contents')
plt.xticks(rotation=45)
plt.show()
```
